```python
import pandas as pd
```

```python
# Importation des avertissements et suppression des avertissements pour une meilleure lisibilité
import warnings
warnings.filterwarnings("ignore")

# Importation de la bibliothèque pandas et lecture du fichier CSV 'ADNI_Gene_Expression_Profile.csv' dans un DataFrame
df = pd.read_csv('ADNI_Gene_Expression_Profile.csv')
```

```python
df.head(10)
```

```python
# Définition de la colonne 'Phase' comme index du DataFrame
df = df.set_index('Phase')

# Renommage du nom de l'index à None pour une meilleure lisibilité
df.index.names = [None]

# Affichage des premières lignes du DataFrame après les modifications
df.head()
```

```python
# Transposition du DataFrame, échange des lignes et des colonnes
df = df.T

# Affichage du DataFrame après la transposition
df
```

```python
# Conversion de la colonne 'SubjectID' en type str (chaîne de caractères)
df['SubjectID'] = df['SubjectID'].astype(str)
```

```python
# Insérer les colonnes 'RID' et 'final_diagnosis' avec des valeurs 'nan' (non disponibles) à la position 2 et 3 respectivement
df.insert(2, 'RID', 'nan')
df.insert(3, 'final_diagnosis', 'nan')

# Afficher les premières lignes du DataFrame après l'insertion des colonnes
df.head()
```

```python
# Lecture du fichier CSV 'adni_pts_info.csv' dans un nouveau DataFrame
df_pheno = pd.read_csv('adni_pts_info.csv')

# Affichage du DataFrame contenant les informations phénotypiques
df_pheno
```

```python
# Conversion de la colonne 'RID' en type str (chaîne de caractères)
df_pheno['RID'] = df_pheno['RID'].astype(str)
```

```python
# Ajout d'une nouvelle colonne 'Length' contenant la longueur de chaque valeur dans la colonne 'RID'
df_pheno["Length"] = df_pheno["RID"].str.len()

# Affichage du DataFrame avec la colonne 'Length' ajoutée
df_pheno
```

```python
# Parcourir chaque ligne du DataFrame
for i in range(len(df_pheno)):
    # Vérifier la longueur de l'identifiant RID
    if df_pheno['Length'][i] == 1:
        # Si la longueur est 1, ajouter trois zéros à gauche
        df_pheno['RID'][i] = '000' + df_pheno['RID'][i]
    elif df_pheno['Length'][i] == 2:
        # Si la longueur est 2, ajouter deux zéros à gauche
        df_pheno['RID'][i] = '00' + df_pheno['RID'][i]
    elif df_pheno['Length'][i] == 3:
        # Si la longueur est 3, ajouter un zéro à gauche
        df_pheno['RID'][i] = '0' + df_pheno['RID'][i]
```

```python
df_pheno
```

```python
# Réinitialisation de l'index du DataFrame
df = df.reset_index()

# Affichage des premières lignes du DataFrame après la réinitialisation de l'index
df.head()
```

```python
# Parcourir chaque ligne du DataFrame 'df'
for i in range(len(df)):
    # Parcourir chaque ligne du DataFrame 'df_pheno'
    for j in range(len(df_pheno)):
        # Vérifier si l'identifiant RID dans 'df_pheno' est présent dans 'SubjectID' de 'df'
        if df_pheno['RID'][j] in df['SubjectID'][i]:
            # Si c'est le cas, associer le RID et le diagnostic final correspondants à l'identifiant de sujet dans 'df'
            df['RID'][i] = df_pheno['RID'][j]
            df['final_diagnosis'][i] = df_pheno['final_diagnosis'][j]
```
